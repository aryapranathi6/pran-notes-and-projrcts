````python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
````

````python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
````

````python
## to do
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame

# 1. Extract numeric features for PCA
numeric_features = data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = data[numeric_features]

# 2. Standardize the data (important for PCA)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# 3. Perform PCA
pca = PCA()  # Initialize PCA without specifying n_components to keep all
pca.fit(scaled_data)

# 4. Calculate and plot the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o')
plt.title('Cumulative Explained Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# 5. Determine the optimal number of components
# You can choose the number of components that explain a desired amount of variance (e.g., 95%)
desired_variance = 0.95  # Set your desired explained variance threshold
optimal_n_components = np.argmax(cumulative_variance_ratio >= desired_variance) + 1

print(f"Optimal number of principal components: {optimal_n_components}")
````

````python
## to do
# visualise the principal components, choose the number of principal components based on the above plot. What is you observation from the plot?


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Assuming 'data' is your DataFrame and you've determined optimal_n_components

# 1. Extract numeric features and standardize
numeric_features = data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = data[numeric_features]
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_data)

# 2. Perform PCA with the chosen number of components
pca = PCA(n_components=2)  # Replace 2 with your optimal_n_components
pca_data = pca.fit_transform(scaled_data)

# 3. Create a DataFrame with the principal components and target variable ('Tea' in this case)
pca_df = pd.DataFrame(data=pca_data, columns=['PC1', 'PC2'])
pca_df['Tea'] = data['Tea']

# 4. Visualize the principal components using a scatter plot
sns.scatterplot(x='PC1', y='PC2', hue='Tea', data=pca_df)
plt.title('PCA Visualization of Starbucks Nutrition Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
````

````python
## to do
# plot the first 2 components of tsne, whats you observation from the plot?

from sklearn.manifold import TSNE
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and 'Tea' is your target variable

# 1. Extract numeric features
numeric_features = data.select_dtypes(include=['float64', 'int64']).columns
numeric_data = data[numeric_features]

# 2. Perform t-SNE
tsne = TSNE(n_components=2, random_state=42)  # Set random_state for reproducibility
tsne_data = tsne.fit_transform(numeric_data)

# 3. Create DataFrame with t-SNE components and target variable
tsne_df = pd.DataFrame(data=tsne_data, columns=['t-SNE 1', 't-SNE 2'])
tsne_df['Tea'] = data['Tea']

# 4. Visualize the t-SNE components using a scatter plot
sns.scatterplot(x='t-SNE 1', y='t-SNE 2', hue='Tea', data=tsne_df)
plt.title('t-SNE Visualization of Starbucks Nutrition Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
````
